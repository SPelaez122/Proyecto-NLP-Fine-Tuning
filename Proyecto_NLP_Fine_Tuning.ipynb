#Descripcion de los datasets y los problemas-tareas asociados



*   **Sequence Classification:** El dataset que escogi llamado AI-human-text, tiene textos como entrada, y el problema sera clasificarlos en dos simples clases, si este fue escrito por un humano o una IA, por lo tanto el output sera binario, 0 para humano, 1 para IA. Tiene mas de 400k registros.

*   **Sequence to sequence:** Utilizare el dataset llamado arxiv-summarization, tiene papers del repositorio arxiv como entrada , y el problema sera hacer un resumen de este paper, como un tipo abstract, muy importante en este tipo de publicaciones. El output es lo que se esperaria que se resuma.

### Setting up Tokenizers and Models for Fine-tuning

#### 1. Sequence Classification (AI-human-text Dataset) - using DistilBERT

```python
id2label = {0: "HUMAN", 1: "AI"}
label2id = {"HUMAN": 0, "AI": 1}
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# Model checkpoint for sequence classification
model_checkpoint_ai_human = "distilbert-base-uncased"

# For AI-human classification (0 for human, 1 for AI)
num_labels_ai_human = 2

# Load the tokenizer
tokenizer_ai_human = AutoTokenizer.from_pretrained(model_checkpoint_ai_human)

# Load the model with a classification head
model_ai_human = AutoModelForSequenceClassification.from_pretrained(
    model_checkpoint_ai_human,
    num_labels=num_labels_ai_human,id2label=id2label, label2id=label2id
)

print(f"Tokenizer loaded for AI-human classification: {model_checkpoint_ai_human}")
print(f"Model for Sequence Classification loaded: {model_checkpoint_ai_human} with {num_labels_ai_human} labels")
```

#### 2. Sequence Labeling (ner_movie Dataset) - using DistilBERT

```python
# Extract all unique NER tags from the 'label' column
all_labels_str = ner_movie['train']['label']
unique_labels = set()
for label_str in all_labels_str:
    # Split the comma-separated string and add to the set
    for label in label_str.split(','):
        unique_labels.add(label.strip())

# Sort the unique labels to ensure consistent mapping
label_names_ner_movie = sorted(list(unique_labels))

# Create id2label and label2id mappings
id2label_ner = {i: label for i, label in enumerate(label_names_ner_movie)}
label2id_ner = {label: i for i, label in enumerate(label_names_ner_movie)}

print("id2label for NER:", id2label_ner)
print("label2id for NER:", label2id_ner)
print(f"Number of labels for NER: {len(label_names_ner_movie)}")
```

```python
import numpy as np
import evaluate

# Load the metrics globally
accuracy_metric = evaluate.load("accuracy")
f1_metric = evaluate.load("f1")
seqeval_metric = evaluate.load("seqeval")
rouge_metric = evaluate.load("rouge") # Load ROUGE metric for summarization
```

```python
def compute_metrics_classification(eval_pred):
    predictions, labels = eval_pred.predictions, eval_pred.label_ids
    if isinstance(predictions, tuple):
        predictions = predictions[0]
    predictions = np.argmax(predictions, axis=1)
    accuracy = accuracy_metric.compute(predictions=predictions, references=labels)
    f1 = f1_metric.compute(predictions=predictions, references=labels)
    return {"accuracy": accuracy["accuracy"], "f1": f1["f1"]}
```

```python
def compute_metrics_ner(eval_pred):
    predictions, labels = eval_pred.predictions, eval_pred.label_ids
    if isinstance(predictions, tuple):
        predictions = predictions[0]
    predictions = np.argmax(predictions, axis=2) # Get predicted label ID for each token

    true_labels = [[id2label_ner[l] for l in label_list if l != -100] for label_list in labels]
    true_predictions = [[id2label_ner[p] for (p, l) in zip(pred_list, label_list) if l != -100] for pred_list, label_list in zip(predictions, labels)]

    results = seqeval_metric.compute(predictions=true_predictions, references=true_labels)

    return {
        "precision": results["overall_precision"],
        "recall": results["overall_recall"],
        "f1": results["overall_f1"],
        "accuracy": results["overall_accuracy"],
    }
```

```python
def compute_metrics_summarization(eval_pred):
    predictions, labels = eval_pred.predictions, eval_pred.label_ids
    if isinstance(predictions, tuple):
        predictions = predictions[0]

    # Convert logits (3D array) to predicted token IDs (2D array)
    predictions = np.argmax(predictions, axis=-1)

    decoded_preds = tokenizer_summarization.batch_decode(predictions, skip_special_tokens=True)
    labels = np.where(labels != -100, labels, tokenizer_summarization.pad_token_id)
    decoded_labels = tokenizer_summarization.batch_decode(labels, skip_special_tokens=True)

    result = rouge_metric.compute(predictions=decoded_preds, references=decoded_labels, use_stemmer=True)
    result = {key: value * 100 for key, value in result.items()}
    return result
```

### Tokenizing Datasets

```python
def tokenize_ai_human_function(examples):
    tokenized_inputs = tokenizer_ai_human(examples["text"], truncation=True)
    # Add the labels (the 'generated' column) to the tokenized inputs
    tokenized_inputs["labels"] = examples["generated"]
    return tokenized_inputs

tokenized_ai_human = ai_to_human.map(tokenize_ai_human_function, batched=True)

print("Tokenized AI-human-text dataset example:")
print(tokenized_ai_human["train"][0])
```

```python
def tokenize_and_align_labels(examples):
    tokenized_inputs = tokenizer_ai_human(
        examples["text"],
        truncation=True,
    )

    labels = []
    for i, label_str in enumerate(examples["label"]):
        word_ids = tokenized_inputs.word_ids(batch_index=i)
        previous_word_idx = None
        label_ids = []
        # Convert string labels to integer IDs
        example_labels_ids = [label2id_ner[l.strip()] for l in label_str.split(',')]

        for word_idx in word_ids:
            if word_idx is None:
                label_ids.append(-100)
            elif word_idx != previous_word_idx:
                if word_idx < len(example_labels_ids): # Use example_labels_ids here
                    label_ids.append(example_labels_ids[word_idx])
                else:
                    # Fallback for cases where word_idx exceeds available example_labels_ids
                    # This could happen if the number of words from word_ids is more than the number of explicit labels.
                    label_ids.append(-100)
            else:
                label_ids.append(-100)
            previous_word_idx = word_idx

        labels.append(label_ids)

    tokenized_inputs["labels"] = labels
    return tokenized_inputs

# Apply the preprocessing function and remove original columns to ensure only tokenized data remains
tokenized_ner_movie = ner_movie.map(tokenize_and_align_labels, batched=True, remove_columns=ner_movie["train"].column_names)

print("Tokenized ner_movie dataset example (first example, showing all features):")
print(tokenized_ner_movie["train"][0])
```

```python
def preprocess_function(examples):
    # Truncate article to 512 words before adding prefix and tokenizing
    truncated_articles = [" ".join(doc.split(' ')[:512]) for doc in examples["article"]]
    inputs = ["summarize: " + doc for doc in truncated_articles]
    model_inputs = tokenizer_summarization(inputs, truncation=True)

    # Truncate abstract to 128 words before tokenizing
    truncated_abstracts = [" ".join(abs_doc.split(' ')[:128]) for abs_doc in examples["abstract"]]
    labels = tokenizer_summarization(text_target=truncated_abstracts, truncation=True)

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

# Use the complete arxiv_summarization dataset for tokenization
tokenized_arxiv_summarization_train = arxiv_summarization["train"].map(preprocess_function, batched=True)
tokenized_arxiv_summarization_test = arxiv_summarization["test"].map(preprocess_function, batched=True)

tokenized_arxiv_summarization = {
    "train": tokenized_arxiv_summarization_train,
    "test": tokenized_arxiv_summarization_test
}

print("Tokenized arxiv-summarization dataset example (train split):")
print(tokenized_arxiv_summarization["train"][0])
```

# Training tests

### Data Collators for Each Task

```python
from transformers import DataCollatorWithPadding, DataCollatorForTokenClassification, DataCollatorForSeq2Seq

# Data collator for AI-human text (Sequence Classification)
data_collator_ai_human = DataCollatorWithPadding(tokenizer=tokenizer_ai_human)
print("Data collator for AI-human text (sequence classification) created.")

# Data collator for NER movie (Token Classification)
data_collator_ner_movie = DataCollatorForTokenClassification(tokenizer=tokenizer_ai_human)
print("Data collator for NER movie (token classification) created.")

# Data collator for Arxiv Summarization (Sequence to Sequence)
data_collator_summarization = DataCollatorForSeq2Seq(tokenizer=tokenizer_summarization, model=model_summarization)
print("Data collator for Arxiv Summarization (sequence to sequence) created.")
```

### Hyperparameter Tuning Tests on Sampled Data

#### Task 1: AI-human-text (Sequence Classification)

```python
from transformers import TrainingArguments, Trainer

# Sample a small subset of the tokenized AI-human dataset
sampled_ai_human_train = tokenized_ai_human["train"].shuffle(seed=42).select(range(1000))
sampled_ai_human_eval = tokenized_ai_human["test"].shuffle(seed=42).select(range(1000))

print(f"Sampled AI-human train dataset size: {len(sampled_ai_human_train)}")
print(f"Sampled AI-human eval dataset size: {len(sampled_ai_human_eval)}")

# Define a list of learning rates to test
learning_rates = [2e-5, 2e-4, 2e-3, 2e-2, 2e-1]

for lr in learning_rates:
    print(f"\n--- Starting training for AI-human text classification with learning rate: {lr} ---")

    # Define TrainingArguments for AI-human text with the current learning rate
    training_args_ai_human = TrainingArguments(
        output_dir=f"./results_ai_human_classification_lr_{lr}", # Unique output directory for each LR
        learning_rate=lr,
        per_device_train_batch_size=16,
        per_device_eval_batch_size=16,
        num_train_epochs=2, # Number of training epochs
        weight_decay=0.01,
        eval_strategy="epoch", # Evaluate at the end of each epoch
        save_strategy="epoch", # Save checkpoint at the end of each epoch
        load_best_model_at_end=True, # Load the best model after training
        push_to_hub=False, # Set to True if you want to push to Hugging Face Hub
        report_to="none" # Disable logging to external services for quick tests
    )

    # Initialize Trainer for AI-human text
    trainer_ai_human = Trainer(
        model=model_ai_human, # Re-using the same model instance, which will be reset implicitly or fine-tuned further
        args=training_args_ai_human,
        train_dataset=sampled_ai_human_train,
        eval_dataset=sampled_ai_human_eval,
        data_collator=data_collator_ai_human,
        compute_metrics=compute_metrics_classification,
    )

    # Train the model
    print(f"Starting training for AI-human text classification with LR={lr}...")
    trainer_ai_human.train()
    print(f"Finished training for AI-human text classification with LR={lr}.")

print("\n--- Completed all learning rate experiments for AI-human text classification. ---")
```
